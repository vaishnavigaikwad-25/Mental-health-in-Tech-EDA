# **Project Summary -**

This project explores the “Mental Health in Tech Survey 2014”, which captures self‑reported mental health status, workplace attitudes, and employer support among technology workers worldwide. The dataset includes demographic information such as age, gender, country and state, as well as job‑related variables like company size, remote work status, and whether the employer is a tech company. In addition, it records experiences and perceptions related to mental health, including family history of mental illness, whether the respondent sought treatment, perceived interference of mental health with work, availability of employer benefits, and willingness to discuss issues with coworkers or supervisors.

The primary aim of this exploratory analysis is to understand how mental health conditions and attitudes vary across demographic groups and workplace contexts, and to identify patterns that could inform employer policies. Specific questions include: How common is seeking treatment among tech workers? Does family history relate to treatment‑seeking? How do employer‑provided benefits and communication about mental health influence employees’ comfort and behaviour?

The analysis follows a structured EDA approach. First, the dataset is inspected for structure, missing values, duplicates, and outliers, with a particular focus on cleaning inconsistencies in variables such as age and gender. Next, univariate analysis summarizes distributions of key variables like treatment, work interference, benefits, and company size. Bivariate and multivariate visualizations then explore relationships between treatment and predictors such as family history, gender, geography, remote work, and employer practices around mental health benefits and wellness programs

Preliminary results suggest that a substantial proportion of respondents report having sought treatment for a mental health condition, and that treatment‑seeking is more common among those with a family history of mental illness. Many respondents work in organizations that provide some form of mental health benefits, but awareness of care options and wellness programmes is uneven, and perceptions of anonymity and ease of taking medical leave vary widely. Willingness to discuss mental health tends to be higher with close coworkers than with supervisors or potential employers, indicating persistent stigma and perceived professional risk

From a business perspective, these patterns highlight opportunities for employers to improve communication about available resources, strengthen confidentiality, and normalise conversations around mental health. By using EDA to surface such patterns, this project provides evidence‑based insights that can guide HR and leadership teams in designing more supportive policies and cultures for tech workers

## 3. ***Data Wrangling***

### Data Wrangling Code

1. Clean Age

This keeps implausible entries (e.g. 99999999999, 8, etc.) as missing, making summary stats more realistic

```python
# Copy original
df_clean = df.copy()

# Convert Age to numeric and handle invalid ages
df_clean['Age'] = pd.to_numeric(df_clean['Age'], errors='coerce')

# Define reasonable age bounds (e.g. 18–80)
df_clean.loc[(df_clean['Age'] < 18) | (df_clean['Age'] > 80), 'Age'] = np.nan
```

2. Standardise Gender

This adds a clean Gender_simple variable with three categories for analysis and plots.



```python
# Lowercase and strip spaces
df_clean['Gender'] = df_clean['Gender'].str.strip().str.lower()

male_terms = ['male', 'm', 'male-ish', 'maile', 'cis male', 'cis man']
female_terms = ['female', 'f', 'woman', 'cis female', 'female trans', 'femake', 'woman']
# Everything else -> 'other'
def simplify_gender(g):
    if pd.isna(g):
        return np.nan
    if g in male_terms:
        return 'Male'
    if g in female_terms:
        return 'Female'
    return 'Other'

df_clean['Gender_simple'] = df_clean['Gender'].apply(simplify_gender)
```

3. Handle key missing categorical values

These replacements make categorical plots easier because they avoid dropping rows with NaN in important variables

```python
# Fill state only for non-US? leave as NaN, else 'Unknown'
df_clean['state'] = df_clean['state'].fillna('Unknown')

# self_employed: replace NaN with 'No' or 'Unknown' (choose and justify)
df_clean['self_employed'] = df_clean['self_employed'].fillna('Unknown')

# work_interfere: NaN -> 'Unknown'
df_clean['work_interfere'] = df_clean['work_interfere'].fillna('Unknown')
```

### What all manipulations have you done and insights you found?

During data wrangling, the original dataset was copied into a working dataframe and the Age variable was converted to numeric with unrealistic values outside 18–80 set to missing. The free‑text Gender field was normalised to lower case and mapped into a new three‑level variable (Male, Female, Other) to reduce noise from spelling variants and diverse labels. Missing values in state, self_employed, and work_interfere were recoded to explicit categories such as “Unknown” to retain observations while still flagging uncertainty for later analysis

#### Chart - 12 - Proportions of treatment within each work_interfere category

```python
# Compute proportions of treatment within each work_interfere category
order = ['Never', 'Rarely', 'Sometimes', 'Often', 'Unknown']
temp = (
    df_clean.groupby(['work_interfere', 'treatment'])
    .size()
    .reset_index(name='count')
)

# Normalize within each work_interfere level using transform for index alignment
temp['prop'] = temp.groupby('work_interfere')['count'].transform(lambda x: x / x.sum())

# Keep ordered categories that exist
order = [x for x in order if x in temp['work_interfere'].unique()]

plt.figure(figsize=(8, 4))
sns.barplot(
    x='work_interfere',
    y='prop',
    hue='treatment',
    data=temp,
    order=order
)
plt.title('Proportion of treatment by work interference level')
plt.xlabel('How much mental health interferes with work')
plt.ylabel('Proportion within each level')
plt.xticks(rotation=20)
plt.legend(title='Treatment')
plt.show()
```

##### 1. Why did you pick the specific chart?

A proportion bar chart shows, within each level of work interference, what fraction of respondents have sought treatment, giving a clearer multivariate view than raw counts alone.
​



##### 2. What is/are the insight(s) found from the chart?

As interference moves from “Never” to “Often”, the share of respondents reporting treatment typically increases, suggesting that greater functional impact is associated with higher treatment uptake.

Clear, repeated communication about available benefits and care options, strong anonymity guarantees, and simple, supportive leave procedures can lower barriers to seeking help and reduce productivity losses from untreated conditions.


# **Conclusion**

This EDA highlights that mental health issues are widespread in the tech workforce and that many employees have either sought treatment or experience some level of work interference, while a substantial share still feel employers do not treat mental and physical health equally. Patterns around benefits, wellness programmes, anonymity, and observed consequences suggest that supportive, well‑communicated policies encourage help‑seeking, whereas uncertainty or perceived stigma can discourage early intervention. By acting on these insights—improving policies, manager behaviour, and communication—tech organisations can create a more supportive culture that protects employee well‑being and sustains long‑term performance.
